```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("../input/sales-data/sales_data.csv");
```

```python
print(df.head())
print(type(df.columns))
```

```python
plt.xlabel("Month Number");
plt.ylabel("Total Profit");
plt.plot(df['month_number'], df['total_profit'],"g--o", linewidth = 4, markerfacecolor = 'k', markersize = 10);
```

```python
products = [c for c in df.columns]

products.pop(0)
products.pop(-1)
products.pop(-1)

print(products)

for x in products:
    plt.plot(df['month_number'], "o", df[x], label = x)

plt.show()
```

```python
plt.scatter(df['month_number'], df['toothpaste'])
plt.grid()
plt.xlabel("Month Number")
plt.ylabel("Number of units of Toothpaste")
```

```python
months = [i + 1 for i in range(12)]
facecream_bar = np.arange(len(df['facecream']))
facewash_bar = [x + 0.4 for x in facecream_bar]

plt.figure(figsize = (10, 10))
plt.xticks(facecream_bar + 0.2, months)
plt.bar(facecream_bar, df['facecream'], width = 0.4, label = 'Facecream')
plt.bar(facewash_bar, df['facewash'], width = 0.4, label = 'Facewash')
plt.grid()
plt.legend()
```

```python
months = [i + 1 for i in range(12)]
x_bar = np.arange(len(df['bathingsoap']))
plt.figure(figsize = (10, 10))
plt.xticks(x_bar, months)
plt.legend()
plt.bar(x_bar, df['bathingsoap'], label = "Bathingsoap")

plt.savefig('bathingsoap.png')
```

```python
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.figure(figsize = (10, 10))
plt.hist(df['total_profit'], bins = profit_range, label = "Profit Data")
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit Data')
plt.show()
```

```python
products_bought = []

for x in products:
    products_bought.append(df[x].sum())

plt.figure(figsize = (10, 10))
plt.pie(products_bought, labels = products, autopct='%1.2f%%')
plt.legend()
plt.show()
```

```python
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(df ['month_number'], df ['bathingsoap'], label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(df ['month_number'], df ['facewash'], label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')

plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

```python
monthList  = df ['month_number'].tolist()

faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

plt.figure(figsize = (10, 10))

plt.stackplot(monthList, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['m','c','r','k','g','y'])

plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(products, loc='upper left')
plt.show()
```
